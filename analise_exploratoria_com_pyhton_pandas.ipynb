```python
#como filtrar os dados, primeiro √© necess√°rio filtrar quais s√£o os itens unicos.
df["SubCategoryID"].unique()
```

```python
#como fazer agrupamento de dados, neste exemplos vou verificar quantos produtos est√£o listados na sub categoria igual 8
df.groupby("SubCategoryID")["ProductName"].nunique()
```

```python
#como saber a m√©dia
df["ListPrice"].mean()
```

```python
#realizar soma
df['StandardCost'].sum()
```

```python
#Alterando o tipo de dado da coluna
df["SubCategoryID"] = df["SubCategoryID"].astype("object")
df.dtypes
```

```python
#consultando para saber se existe algum valor nulo ou faltando
df.isnull().sum()
```

```python
#caso tenha algum valor nulo ou faltando voc√™ pode substituir os valores nulos por zero ou pela m√©dia, mas se necess√°rio voc√™ tamb√©m pode apagar.
#para substituir pela m√©dia basta usar: df["nome da coluna"].fillna(df["nome da coluna"].mean(), inplace=True)
#para substituir por 0: df["nome da coluna"].fillna(0, inplace=true)
#para apagar: df.dropna(inplace = True)
```

```python
#como criar uma nova coluna
df["receita"] = df["ListPrice"].mul(df["SubCategoryID"])
df.head()
```

```python
df["qtdvendas"] = df["receita"] / df["ListPrice"]
df.head()
```

```python
#como saber qual √© o maior resultado de algo
df["receita"].max()
```

```python
#como saber qual √© o menor resultado de algo
df["receita"].min()
```

```python
#transformando os valores em int
df["receita"] = df["receita"].astype("int64")
df["qtdvendas"] = df["qtdvendas"].astype("int64")
df.dtypes
```

```python
#retorna o top 3 maior
df.nlargest(3,"receita")
```

```python
#retorna o top 3 menor
df.nsmallest(3, "receita")
```

```python
#saber qual foi os produtos que mais venderam
df.groupby("SubCategoryID")["receita"].sum()
```

```python
#ordenando do maior para o menor
df.sort_values("receita", ascending=False).head(20)
```

```python
df["SubCategoryID"].value_counts(ascending=False)
```
